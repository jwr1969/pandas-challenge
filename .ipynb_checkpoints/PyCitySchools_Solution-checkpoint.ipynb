```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# number of unique school names
school_count = school_data_complete["school_name"].nunique()
school_count
```

```python
# Number of math passes
math_passes = school_data_complete.loc[school_data_complete["math_score"] >=70]["Student ID"].count()
# As a percentage of total students
math_passes = 100*math_passes/student_count
math_passes
```

```python
# Number of reading passes
reading_passes = school_data_complete.loc[school_data_complete["reading_score"] >=70]["Student ID"].count()
# As a percentage of total students
reading_passes/student_count
reading_passes = 100*reading_passes/student_count
reading_passes
```

```python
# Number of overall passes
overall_passes = school_data_complete.loc[(school_data_complete["reading_score"] >=70) & 
                                          (school_data_complete["math_score"] >=70) ]["Student ID"].count()
# As a percentage of total students
overall_passes = 100*overall_passes/student_count
overall_passes
```

```python
# Create a dataframe to hold results
Summary_Table_1 = pd.DataFrame({"Schools": school_count, "Students": student_count, "Budget": budget,
                                "Average_Math_Score": avg_math_score, "Average_Reading_Score": avg_reading_score,
                                "Math_Passes%": math_passes, "%Reading_Passes": reading_passes, 
                                "%Overall_Passes":overall_passes}, index = [0])
# Format values
Summary_Table_1["Students"] = Summary_Table_1["Students"].map("{:,}".format)
Summary_Table_1["Students"] = Summary_Table_1["Budget"].map("${:,}".format)
# Display final result
Summary_Table_1
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# School names
school_name = school_data_complete.groupby("school_name").count()
# School type
school_type = school_data_complete.groupby("school_name")["type"].first()
# Student count by school
student_count_byschool = school_data_complete.groupby("school_name")["Student ID"].count()
# Budget by school
budget_byschool = school_data_complete.groupby("school_name")["budget"].first()
# Budget per student by school
budgetperstudent_byschool = budget_byschool/student_count_byschool
# Average math score by school
avg_mathscore_byschool = school_data_complete.groupby("school_name")["math_score"].mean()
# Average reading score by school
avg_readingscore_byschool = school_data_complete.groupby("school_name")["reading_score"].mean()
# % students passing math byschool
mathpasses_byschool = 100*school_data_complete[lambda x:x.math_score >=70].groupby("school_name")["Student ID"].count()/student_count_byschool
# % students passing reading byschool
readingpasses_byschool = 100*school_data_complete[lambda x:x.reading_score >=70].groupby("school_name")["Student ID"].count()/student_count_byschool
# % students passing overall byschool
overallpasses_byschool = 100*school_data_complete[lambda x:(x.reading_score >=70) & (x.math_score >=70)].groupby("school_name")["Student ID"].count()/student_count_byschool
```

```python
# Create a dataframe to hold results
Summary_Table_2 = pd.DataFrame({"School Type": school_type, "Total Students":student_count_byschool, 
                                "Total School Budget": budget_byschool,"Per Student Budget": budgetperstudent_byschool, 
                                "Average Math Score": avg_mathscore_byschool, "Average Reading Score":avg_readingscore_byschool, 
                                "%Passing Math":mathpasses_byschool, "%Passing Reading": readingpasses_byschool, 
                                "%Overall Passing":overallpasses_byschool})
# Format values
Summary_Table_2["Total Students"] = Summary_Table_2["Total Students"].map("{:,}".format)
Summary_Table_2["Total School Budget"] = Summary_Table_2["Total School Budget"].map("${:,}".format)
Summary_Table_2["Per Student Budget"] = Summary_Table_2["Per Student Budget"].map("${:,}".format)
# Display results
Summary_Table_2
```

```python
# Use previous view and sort values in overall passing in descending order and display top 5 results
Summary_Table_2.sort_values(["%Overall Passing"],ascending=False).head(5)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Add columns for budget per student (school budget / school size)
school_data_complete["budget per student"] = school_data_complete["budget"]/school_data_complete["size"]
school_data_complete.head()
```

```python
# Perfoming binning - create categories of budget per student
bins = [0, 584.5, 629.5, 645.5, 700]
group_names = ["<$584","$585-629","$630-644","$645-675"]
# Cut data
school_data_complete["budget per student categories"] = pd.cut(school_data_complete["budget per student"], bins, 
                                                               labels=group_names, include_lowest=True)
```

```python
# Calculate average math and reading scores by budget category
avg_math_score_bybudgetcat = school_data_complete.groupby("budget per student categories")["math_score"].mean()
avg_reading_score_bybudgetcat = school_data_complete.groupby("budget per student categories")["reading_score"].mean()
```

```python
# Calculate number of students in each category
students_per_bybudgetcat = school_data_complete.groupby("budget per student categories")["Student ID"].count()
# Calculate math, reading and overall passes by budget category
math_passes_bybudgetcat = 100*school_data_complete[lambda x:x.math_score >=70].groupby("budget per student categories")["Student ID"].count()/students_per_bybudgetcat
reading_passes_bybudgetcat = 100*school_data_complete[lambda x:x.reading_score >=70].groupby("budget per student categories")["Student ID"].count()/students_per_bybudgetcat
overall_passes_bybudgetcat = 100*school_data_complete[lambda x:(x.reading_score >=70) & (x.math_score >=70)].groupby("budget per student categories")["Student ID"].count()/students_per_bybudgetcat
```

```python
# Create a summary table to hold results
Summary_Table_3 = pd.DataFrame({"Average Math Score": avg_math_score_bybudgetcat,
                                "Average Reading Score": avg_reading_score_bybudgetcat, 
                                "% Passing Math": math_passes_bybudgetcat, 
                                "% Passing Reading": reading_passes_bybudgetcat, 
                                "% Overall Pass": overall_passes_bybudgetcat})
# format and display results
Summary_Table_3.applymap("{:,.4}".format)
```
